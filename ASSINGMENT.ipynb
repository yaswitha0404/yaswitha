Code  for Natural growth and decay

```python
import math
import matplotlib.pyplot as plt

def exponential_growth(initial_value, growth_rate, time_interval, total_time):
    values = [initial_value]
    time = 0
    while time < total_time:
        value_change = growth_rate * values[-1] * time_interval
        values.append(values[-1] + value_change)
        time += time_interval
    return values

def exponential_decay(initial_value, decay_rate, time_interval, total_time):
    values = [initial_value]
    time = 0
    while time < total_time:
        value_change = decay_rate * values[-1] * time_interval
        values.append(values[-1] - value_change)
        time += time_interval
    return values

def plot_growth_or_decay(values, time_interval, title):
    time_points = [i * time_interval for i in range(len(values))]
    plt.plot(time_points, values)
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title(title)
    plt.show()

initial_value = 80
growth_rate = 0.04
decay_rate = 0.04
time_interval = 1
total_time = 15


growth_values = exponential_growth(initial_value, growth_rate, time_interval, total_time)
plot_growth_or_decay(growth_values, time_interval, "Exponential Growth")

decay_values = exponential_decay(initial_value, decay_rate, time_interval, total_time)
plot_growth_or_decay(decay_values, time_interval, "Exponential Decay")
```
